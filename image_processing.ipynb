```python
def print_image_details(name, image_data):
    print(f"--- {name} ---")
    print(f"Shape: {image_data.shape}")
    print(f"Data type: {image_data.dtype}")
    if image_data.ndim == 3:
        print(f"Color order assumption: {'BGR (OpenCV default)' if name.startswith('OpenCV') else 'RGB (Matplotlib/Scikit-image default)'}")
    print("-" * (len(name) + 8))
img_cv2 = cv2.imread('download.jpg', cv2.IMREAD_COLOR)
if img_cv2 is None:
    print("Error: download.jpg not found or could not be read. Please ensure the image is in the correct path.")
else:
    print_image_details("OpenCV Image", img_cv2)
    img_mpl = plt.imread('download.jpg')
    print_image_details("Matplotlib Image", img_mpl)
    img_skimage = io.imread('download.jpg')
    print_image_details("Scikit-image Image", img_skimage)
    fig, axes = plt.subplots(1, 3, figsize=(15, 5))
    axes[0].imshow(img_cv2)
    axes[0].set_title("OpenCV (BGR)")
    axes[0].axis('off')
    axes[1].imshow(img_mpl)
    axes[1].set_title("Matplotlib (RGB)")
    axes[1].axis('off')
    axes[2].imshow(img_skimage)
    axes[2].set_title("Scikit-image (RGB)")
    axes[2].axis('off')
    plt.show()
```

```python
img_bgr = cv2.imread("download.jpg")
img_rgb = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB)
R = img_rgb[:, :, 0]
G = img_rgb[:, :, 1]
B = img_rgb[:, :, 2]
swapped = img_rgb.copy()
swapped[:, :, 0] = B  
plt.figure(figsize=(12,4))
plt.subplot(1,3,1)
plt.title("Original")
plt.imshow(img_rgb)
plt.axis("off")
plt.subplot(1,3,2)
plt.title("Blue Channel")
plt.imshow(B, cmap="gray")
plt.axis("off")
plt.subplot(1,3,3)
plt.title("Swapped (R  B)")
plt.imshow(swapped)
plt.axis("off")
plt.show()
```
